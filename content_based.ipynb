```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/movies.csv")     #Movies dataset
user = pd.read_csv("data/ratings.csv")  #Rating dataset
```

Find all possible genres in the movies datasets.

```python
genres = []
for i in range(len(df["genres"])):
    current = df["genres"][i].split("|")
    for j in current:
        if j not in genres and j != '(no genres listed)':
            genres.append(j)
```

```python
dic_genres = {}
for i in range(len(genres)):
    dic_genres[genres[i]] = i
one_hot = [0 for i in range(len(genres))] 
```

```python
one_hots = []
for i in range(len(df["genres"])):
    current_one_hot = one_hot.copy()
    current = df["genres"][i].split("|")
    for j in current:
        if j != '(no genres listed)':
            current_one_hot[dic_genres[j]] = 1
    current_one_hot = np.array(current_one_hot)
    one_hots.append(current_one_hot)
        
```

Add new column to the dataset using one hots that found in the last cell

```python
df["one_hot"] = one_hots
```

Create a function that will calcualte cosine similarity

Create a function that will calculate a likeness list with the user's liked movies

```python
def similarity(list1, list2):
    return np.dot(list1, list2) / np.sqrt(np.sum(list1) * np.sum(list2))



def get_one_hot(ids, dataframe):
    one_hot = [0 for x in range(19)]
    for id in ids:

        current = dataframe[dataframe["movieId" ] == id]
        current = list(current["one_hot"])[0]

        for j in range(19):
            one_hot[j] += current[j]


    for i in range(len(one_hot)):
        one_hot[i] = one_hot[i] / len(ids)
          
    return one_hot

```

Create a accuracy test, that will calculate accuracy using test movies and predicted movies

```python
def accuracy(test, predictions):
    correct = 0
    for prediction in predictions:
        if prediction in test:
            correct += 1
    return (correct/len(test))*100
```

Main function that will do the work of prediction

```python
def content(train, df, ids, test):
    like_degree = get_one_hot(train, df)

    sims = []
    for i in range(len(df)):
        sims.append(similarity(like_degree, df["one_hot"][i]))
    sims = np.array(sims)
    indices = np.argsort(sims)[-len(test):][::-1]
    #indices = sorted(range(len(sims)), key=lambda i: sims[i], reverse=True)[:(5+len(train)) ]

    result = []
    for index in indices:
        if df["movieId"][index] not in ids:
            result.append(df["movieId"][index])

    result = accuracy(test, result)
    return result

    
```

A loop which will try every user's id and recommend best movies for them.

```python
from sklearn.utils import shuffle

results = []
for id in range(1,611):
    user_id = id
    user_movies = user[user["userId"] == user_id]
    mean = user_movies["rating"].mean()

    user_liked = user_movies[user_movies["rating"] >= mean]

    ids = []

    for row in user_liked["movieId"]:
        ids.append(row)
    ids = shuffle(ids)
    point = int(0.8 * len(ids))
    train = ids[:point]
    test = ids[point:]
    result = content(train, df, ids, test)
    results.append(result)

np.mean(results)
```
